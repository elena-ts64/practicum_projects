# Предсказание увольнения сотрудников из компании.

**Описание исследования.**

Исследование необходимо для HR отдела компании «Работа с заботой» для предсказания оттока сотрудников и минимизации рисков компании, связанных с увольнениями. 

Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли. Поэтому данный показатель - уровень удовлетворённости сотрудника - необходимо предсказывать с помощью машинного обучения. Почему бизнесу это важно: удовлетворённость работой напрямую влияет на отток сотрудников. А предсказание оттока — одна из важнейших задач HR-аналитиков. Внезапные увольнения несут в себе риски для компании, особенно если уходит важный сотрудник. Вторая задача — построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.

**Цель исследования.**

Минимизировать финансовые потери компании, связанные с оттоком сотрудников.

**Задачи исследования.**
1. Построить модель машинного обучения, которая сможет на основе данных заказчика предсказать уровень удовлетворённости сотрудника, необходимого для предсказания оттока.
2. Построить модель машинного обучения, которая сможет на основе данных заказчика и пердсазанного уровня удовлетворённости сотрудника предсказать увольнение сотрудника из компании.

```python
def space_values_deletion(df):
    df = df.replace(r'^\s*$', np.nan, regex=True)
    return df

def remove_typo(df):
    try:
        df['level'] = df['level'].replace('sinior', 'senior')
        return df
    except:
        return df

gaps_processing = Pipeline([
    ('space_values_deletion', FunctionTransformer(space_values_deletion)),
    ('remove_typo', FunctionTransformer(remove_typo))
])
```

```python
def plot_box_hist(param, df):
    #построение ящика с усами
    plt.figure(figsize = (10.5, 2))
    plot_box = sns.boxplot(x=df[param])
    plt.title(f'Разброс значений параметра {param}')
    plt.show()
    #построение гистограммы
    plt.figure(figsize = (10, 6))
    plt.axvline(df[param].mean(), c='red', ls='-', lw=2.5, label='Среднее')
    plt.axvline(df[param].median(), c='green', ls='--', lw=2.5, label='Медиана')
    plt.legend(loc='upper right')
    plot_hist = sns.histplot(data=df, x=df[param])
    plt.title(f'Распределение значений параметра {param}')
    plt.xlabel(f'{param}')
    plt.ylabel('Количество')
    plt.grid(axis='y')
    plt.show()
    return plot_box, plot_hist
```

```python
ohe_pipe = Pipeline(
    [
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ]
)
```

```python
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'],
                    ['low', 'medium', 'high']
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
num_pipe = Pipeline(
    [
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('scaler', StandardScaler())
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', num_pipe, num_columns)
    ],
    remainder='passthrough'
)
```

```python
data_preprocessor
```

### Обучение моделей

```python
def metric_smape (y_true, y_pred):
    metric = np.mean((np.abs(y_true - y_pred)) / ((np.abs(y_true) + np.abs(y_pred)) / 2)) * 100
    return metric

smape = make_scorer(metric_smape, greater_is_better=False)
```

```python
pipe_final = Pipeline(
    [
        ('gaps', gaps_processing),
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
)
```

```python
param_grid = [
    {
        'preprocessor__num__imputer':[SimpleImputer(), KNNImputer()],
        'preprocessor__num__scaler': [StandardScaler(), 'passthrough'],
        'models': [LinearRegression()]
    },
    {
        'preprocessor__num__imputer':[SimpleImputer(), KNNImputer()],
        'preprocessor__num__scaler': [StandardScaler(), 'passthrough'],
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(7, 15),
        'models__min_samples_split': range(2, 8)
    },
    {
        'preprocessor__num__imputer':[SimpleImputer(), KNNImputer()],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough'],
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)]
    }
    ,
    {
        'preprocessor__num__imputer':[SimpleImputer(), KNNImputer()],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough'],
        'models': [SGDRegressor()]
    },
    {
        'preprocessor__num__imputer':[SimpleImputer(), KNNImputer()],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough'],
        'models': [SVR()]
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid,
    n_iter=30,
    cv=5,
    scoring=smape,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
start = time.time()
randomized_search.fit(X_train, y_train)
print(f'Время выбора и обучения модели: {(time.time() - start):.3f} секунд')
print()
print('Лучшая модель и её параметры:\n', randomized_search.best_estimator_)
print()
print ('Метрика SMAPE лучшей модели на кросс валидационной выборке:', randomized_search.best_score_ * (-1)) 
# т.к. метрика была создана в make_scorer с параметром greater_is_better=False, 
# итоговое значение домножается на -1 для вывода значения метрики в стандартном виде
```

```python
y_test_pred = randomized_search.predict(X_test)
print(f'Метрика SMAPE на тестовой выборке: {round(metric_smape(y_test, y_test_pred), 3)}')
```
